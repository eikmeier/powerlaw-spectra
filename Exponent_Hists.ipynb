```julia
#non models & non-as, Histogram of statistically significant adj_alpha values
filt = map!(x -> x & !isna(x), data[:adj_p] .>= 0.1) & map!(x -> x & !isna(x), data[:types] .== "AS")
filt1 = map!(x -> x & !isna(x), data[:adj_p] .>= 0.1) & map!(x -> x & !isna(x), data[:types] .== "Facebook")
filt2 = map!(x -> x & !isna(x), data[:adj_p] .>= 0.1) & map!(x -> x & !isna(x), data[:types] .== "Oregon")
filt3 = map!(x -> x & !isna(x), data[:adj_p] .>= 0.1) & map!(x -> x & !isna(x), data[:types] .== "other") &
(data[:subtype].!= "Collab.") & (data[:subtype].!= "Fiction") & (data[:subtype].!= "Tech.") & (data[:subtype].!= "Relational")
LL4 = data[filt,:];
LL1 = data[filt1,:];
LL2 = data[filt2,:];
LL3 = data[filt3,:];

LL = vcat(LL1, LL2, LL3, LL4)

real_adj=Gadfly.plot(LL, x = "adj_alpha", color = "types",Geom.histogram(bincount = 70), 
Theme(grid_strokedash=[]),
Coord.Cartesian(xmin=2, xmax=9, ymax = 12),
Guide.xlabel("Adjacency Powerlaw Î±" ),
Guide.colorkey("Types"))
```
